Fill NA Method

```
import pandas as pd
df = pd.read_csv("weather_data.csv",parse_dates=['day']) #we change the day type from string to date function
type(df.day[0])
df.head()
```

```
df.set_index('day',inplace=True) #if Inplace is fasle so its return new data frame 
df
```

```
new_df = df.fillna(0)
new_df
```

```
# to fill with colum
new_df = df.fillna({
        'temperature': 0,
        'windspeed': 0,
        'event': 'No Event'
    })
new_df
```

```
new_df = df.fillna(method="ffill") #forward fill
new_df
```

```
new_df = df.fillna(method="bfill") #backward fill
new_df
```

```
new_df = df.fillna(method="bfill", axis="columns") # axis is either "index" or "columns" copy thier value
new_df
```

```
new_df = df.fillna(method="ffill",limit=1)
new_df
```

interpoate method

```

new_df = df.interpolate()
new_df
```

Drop Na

```
new_df = df.dropna()
new_df
```

```
new_df = df.dropna(how='all')
new_df
```

```
new_df = df.dropna(thresh=1) 
#treshold represent how much na vaue is allowed tresh=1 means if i have teast one non NaN value keep the row
new_df
```
